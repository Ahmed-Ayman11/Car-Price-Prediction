```python
X_train.shape
```

```python
numerical_features = ['manifacture_year', 'condition', 'milage']
categorical_features = ['brand', 'model', 'trim', 'body']
```

* ## Scaling & Encoding

```python
# Numerical data processing pipeline
numerical_pipeline = Pipeline([
    ('scaler', RobustScaler())  # Scaling numerical data
])

# Categorical data processing pipeline using BinaryEncoder
categorical_pipeline = Pipeline([
    ('binary', ce.BinaryEncoder())  # Binary encoding
])
```

```python
preprocessor = ColumnTransformer([
    ('num', numerical_pipeline, numerical_features),
    ('cat', categorical_pipeline, categorical_features)
])
```

```python
X_train_prep2 = preprocessor.fit_transform(X_train)
X_test_prep2 = preprocessor.transform(X_test)
```

```python
X_train_prep2.shape
```

# Modeling

```python
# Liner Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection  import cross_val_score
from sklearn.metrics import r2_score

lr = LinearRegression()
scores = cross_val_score(lr, X_train_prep2, y_train, cv=5, scoring='r2')
print(scores)
print('Mean R2:', np.mean(scores))
```

```python
from sklearn.metrics import r2_score
```

```python
# Ridge
from sklearn.linear_model import Ridge

ridge = Ridge()
scores = cross_val_score(ridge, X_train_prep2, y_train, cv=5, scoring='r2')
print(scores)
print('Mean R2:', np.mean(scores))
```

```python
# lasso 
from sklearn.linear_model import Lasso

lasso = Lasso()
scores = cross_val_score(lasso, X_train_prep2, y_train, cv=5, scoring='r2')
print(scores)
print('Mean R2:', np.mean(scores))
```

```python
# Desicion Tree
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
scores = cross_val_score(dt, X_train_prep2, y_train, cv=5, scoring='r2')
print(scores)
print('Mean R2:', np.mean(scores))
```

```python
# KNN
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor()
scores = cross_val_score(knn, X_train_prep2, y_train, cv=5, scoring='r2')
print(scores)
print('Mean R2:', np.mean(scores))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
scores = cross_val_score(rf, X_train_prep2, y_train, cv=5, scoring='r2')
print(scores)
print('Mean R2:', np.mean(scores))
```
